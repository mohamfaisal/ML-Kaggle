```python
import gradio as gr

# Simple function
def greet(name):
    return f"Hello, {name}! Welcome to your Gradio interface."

# Create interface
demo = gr.Interface(
    fn=greet,                           # Function to run
    inputs=gr.Textbox(label="Enter your name"),
    outputs=gr.Textbox(label="Greeting"),
    title="Basic Gradio App",
    description="A very simple example interface using Gradio."
)

demo.launch()
```
