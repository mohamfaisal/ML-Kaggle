```python
from textblob import TextBlob
```

```python
# Samples
texts = [
    "I love working with creative teams!",
    "The project is a complete disaster.",
    "This movie was okay, not too bad but not great either.",
]

for text in texts:
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity #ranger from -1(negative) to +1 (positive)

    if polarity > 0:
        sentiment = "positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    print(f"Text: {text}")
    print(f"Sentiment: {sentiment} (Polarity: {polarity:.2f})")
    print("-" * 60)
```
