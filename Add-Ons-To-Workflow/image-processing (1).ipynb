```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('lena.jpg')
print(f"Image Shape: {img.shape}")
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6, 6))
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off') # Hide the x/y axis numbers
plt.show()
```

```python
# 1. Resizing: Make it smaller (100x100 pixels)
img_tiny = cv2.resize(img_rgb, (100, 100))

# 2. Grayscale: Remove color info
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 3. Blurring: Smooth out noise (Gaussian Blur)
# (15, 15) is the kernel size - larger numbers = more blurry
img_blur = cv2.GaussianBlur(img_rgb, (15, 15), 0)

# Display them side by side
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(img_tiny); ax[0].set_title("Resized")
ax[1].imshow(img_gray, cmap='gray'); ax[1].set_title("Grayscale")
ax[2].imshow(img_blur); ax[2].set_title("Blurred")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("/kaggle/working/lena.jpg")

# 1. Convert to Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 2. Apply Canny Edge Detection (use thresholds like 100, 200)
edges = cv2.Canny(gray, 100, 200)

# 3. Display the result
plt.figure(figsize=(8, 8))
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("/kaggle/working/lena.jpg")

# 1. Convert to Grayscale (same as Exercise 2)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 2. Apply Binary Thresholding
# Pixels > 127 → 255 (white)
# Pixels <= 127 → 0 (black)
_, binary_img = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# 3. Display the result
plt.figure(figsize=(8, 8))
plt.imshow(binary_img, cmap='gray')
plt.title("Binary Thresholding")
plt.axis("off")
plt.show()
```
